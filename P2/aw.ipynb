```python
def Hitung_RataRata(durasi):
    n = len(durasi)
    total = 0
    Tmin = durasi[0]
    Tmax = durasi[0]

    for i in range(n):  # i = 0..n-1
        total += durasi[i]

        if durasi[i] < Tmin:
            Tmin = durasi[i]

        if durasi[i] > Tmax:
            Tmax = durasi[i]

    avg = total / n
    return avg, Tmin, Tmax


# Contoh pemanggilan
durasi = [30, 45, 60, 20, 50]
avg, Tmin, Tmax = Hitung_RataRata(durasi)

print("Rata-rata =", avg)
print("Nilai minimum =", Tmin)
print("Nilai maksimum =", Tmax)
```
